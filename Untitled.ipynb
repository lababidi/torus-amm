```python
def invariant(delta, deltain, Y, A, T, T2, ri, n=3,):
    print(delta)
    sqrtn = np.sqrt(n)
    d2 = deltain**2 + delta**2
    ds = (deltain + delta)/sqrtn
    alpha = d2 - 2*A*ds
    beta = d2 - ds*(ds + T)
    gamma = T2 + Y - ri**2 - 2*T*A
    return alpha * (alpha + gamma) - 4*Y*beta


n = 3
sqrtn = np.sqrt(n)
ss_ = (sqrtn / (sqrtn - 1))
MM = 1_000_000
liqs = np.array([8, 1, 1, 1]) * MM
L = liqs.sum()
ri = L * ss_ # true at init because it's just sum of all assets

# prices = np.array([.999, .9991, .9992, .9993, .9994, .9995, .9996,  .9997, .9998, .9999 ])
 # .9993,
prices = np.array([  .9999, .9995, .9993, .999, ])
k_ = (sqrtn - (prices + n -1)/(np.sqrt(n*(prices**2 + n -1)))) 

x_ = np.array([L, L, L])
kb = 0
rb = 0
A = kb + ri*sqrtn
T = np.sum(x_)/sqrtn
T2 = np.sum(x_**2)
Y = rb**2 - (kb - rb*sqrtn)**2
    
swapin = 1037.366596782039
swapin = 1557.01819992

swapout = fsolve(invariant, -swapin, (swapin,Y,A,T,T2,ri))
print("Swap Out ≈", swapout[0])
print("dif", -(swapin + swapout[0]))

xp = x_ + [swapin, swapout[0], 0]
print("total", np.sum(x_)/sqrtn)
print("current", np.sum(xp)/sqrtn)
print("crossover",  k_[0]*ri )
print("crossed", k_[0]*ri - np.sum(xp)/sqrtn)
print(sqrtn*(np.sum(x_)/sqrtn - ( k_[0])*ri ))
```
